```python
#Reading the data in the data.h5 file after the simulation:
#Hoomd schema is used:

rf = "/Users/federico/mnt/ebi/kapsel3/2_squirmers_interaction/OBSTACLE/output/data.h5"
f = h5py.File(rf, 'r')

Lx=128
Ly=128
Lz=128
Np=512 #number of particles in the system
N1=256
N2=Np
N3=0
tmax=100 #number of snapshots chosen in the input file of Kapsel

CM1=np.zeros(((tmax,N1,3)))
CM2=np.zeros(((tmax,N2,3)))
CM=np.zeros(((tmax,Np,3)))
CM1X=np.zeros(N1)
CM1Y=np.zeros(N1)
CM1Z=np.zeros(N1)
CM2X=np.zeros(N2)
CM2Y=np.zeros(N2)
CM2Z=np.zeros(N2)
CM3X=np.zeros(N3)
CM3Y=np.zeros(N3)
CM3Z=np.zeros(N3)
CMX=np.zeros(Np)
CMY=np.zeros(Np)
CMZ=np.zeros(Np)


#To mount a movie of the squirmers in Ovito the position of each particle in the box is requested!
# TO DO: reading the orientations of the particles!!!

def create_frame(i):
        s = gsd.hoomd.Snapshot()
        s.particles.N = Np
        s.configuration.step = i
        s.configuration.box = [Lx-1, Ly-1, Lz-1, 0, 0, 0]
        #s.particles.types = ['0']
        s.particles.types = ['0', '1', '2']
        s.particles.typeid = np.concatenate(([0 for n in range(0,N1)], [1 for n in range(0,N2)], [2 for n in range(0,N3)]))
        CM1[i,:,:]=f['system_data/obstacle/pos'].value 
        CM2[i,:,:]=f['trajectory_data/frame_'+str(i)+'/particle/pos'].value 
        
        #Saving the obstacles
        
        for n in range (0,N1):
            CM1X[n]=CM1[i,n,0]-(Lx-1)*0.5 #If I put Lx it doesn't work for one particle!
            CM1Y[n]=CM1[i,n,1]-(Ly-1)*0.5
            CM1Z[n]=CM1[i,n,2]-(Lz-1)*0.5
        
        
        #Saving the moving particles
        
        for n in range (0,N2):

            CM2X[n]=CM2[i,n,0]-(Lx-1)*0.5 #If I put Lx it doesn't work for one particle!
            CM2Y[n]=CM2[i,n,1]-(Ly-1)*0.5
            CM2Z[n]=CM2[i,n,2]-(Lz-1)*0.5+0.5
            
       
        for n in range (0,N2):

            CM3X[n]=CM2[i,n,0]-(Lx-1)*0.5 #If I put Lx it doesn't work for one particle!
            CM3Y[n]=CM2[i,n,1]-(Ly-1)*0.5
            CM3Z[n]=CM2[i,n,2]-(Lz-1)*0.5-0.5
        
        #Building the arrays of the center of mass of obstacles + squirmers
        
        CMX=np.concatenate(((CM1X,CM2X,CM3X))) 
        CMY=np.concatenate(((CM1Y,CM2Y,CM3Y))) 
        CMZ=np.concatenate(((CM1Z,CM2Z,CM3Z))) 

            
        s.particles.position = [[CMX[n],CMY[n],CMZ[n]] for n in range(Np)]
        return s
    
t = gsd.hoomd.open(name='squirmer_movie.gsd', mode='wb')
t.extend( (create_frame(i) for i in range(tmax)) )
```
